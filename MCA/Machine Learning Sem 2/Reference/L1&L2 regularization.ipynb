```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning/Melbourne_housing_FULL.csv')
```

```python
dataset.nunique()
# returns unique value per feature
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset = dataset[cols_to_use]
```

```python
dataset.shape
```

```python
# checking null values
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']

dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)


dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.shape
```

L1 and L2 must be applied only on numerical values, so convert categorical value to numerical values using one-hot encoder

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha = 50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)

# alpha is the regularization strength teller, or also known as lambda
# tol => tolerance, stop the iterations, if my model error is not reducing after 0.1
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 50, max_iter = 100, tol= 0.1)
ridge_reg.fit(train_X, train_y)
```
