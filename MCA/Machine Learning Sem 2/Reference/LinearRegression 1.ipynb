```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Identify relation bw dependent and independent variable.
From this data set ,we have HeadSize and Brain Weight(grams)

```python
df = pd.read_csv('headbrain.csv')
```

```python
df.head()
```

```python
#independent variable and dependent variable
X = df["Head Size(cm^3)"].values #independent , x is an array of head size
Y = df["Brain Weight(grams)"].values #dependent , y is an array of brain weight
```

```python
# compute correlation coefficient matrix(checks if Linear regression is suitable)
np.corrcoef(X,Y)
# 3 Inferences: Analyse the diagonal , if diagonal is 1 , these 2 var are closely realted .
# The other values are close to 1 , we say that the relation is Positively Related , positive regression ( else if values were negative , then negatively related)
```

```python
#plot the Input data
plt.scatter(X,Y,c ='green',label='Data Points')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend() # what all values have been visualized
plt.show()
```

```python
#mean X and mean Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

#Total number of values
n = len(X)

#using the formula to calculate theta1 and theta2
numer = 0
denom = 0
for i in range(n):
  numer +=(X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)

#printing coefficients
print("Coefficients for regression",b1,b0)
```

```python
y = b0 + b1 * X
```

```python
plt.scatter(X, Y, c='green', label='Data Points')
plt.plot(X, y, color='red', label='Regression Line')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
#SSe calculation
sse = np.sum((Y-y) ** 2)
print("SSE:",sse) # always depends on the variance of the data
```

```python
np.mean(Y)
np.std(Y)
#IT IS AN Okay Model , not excellent
```

```python
#no of samples
n = len(Y)
#RMSE calculation
rmse = np.sqrt(sse / n)
print("RMSE",rmse)
#72 grams , below this , model is working on a given data set
# this is a average model
```
