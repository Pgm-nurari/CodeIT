```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
print(x_train.shape)
print(y_train.shape)
print(y_test.shape)
print(y_test.shape)
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
input_shape = (28,28,1)
num_classes = 10
fashion_model = models.Sequential()

# Layer 1: Conv Layer
fashion_model.add(Conv2D(96,(11,11), strides = 1, padding = 'same', input_shape = input_shape))
fashion_model.add(layers.BatchNormalization()) # regularization must be before activation functoion
fashion_model.add(layers.Activation('relu'))
fashion_model.add(MaxPooling2D(pool_size=(2, 2), strides=(1,1)))

#layer 2: Conv layer
fashion_model.add(Conv2D(256, (5, 5), padding='same'))
fashion_model.add(layers.BatchNormalization())
fashion_model.add(layers.Activation('relu'))
fashion_model.add(MaxPooling2D(pool_size=(2, 2), strides=(1,1)))

# Flatten the data for Fully Connected Layers
fashion_model.add(Flatten())

# Layer 8 : Fully Connected Layer
fashion_model.add(Dense(num_classes))
fashion_model.add(layers.Activation('softmax'))

```

```python
fashion_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.01), metrics=['accuracy'])
```

```python
model_history = fashion_model.summary()
```

```python
history = fashion_model.fit(x_train, y_train, batch_size = 512, epochs = 2,
                            validation_data = (x_test, y_test),
                            verbose = 1)
```
